(a) Gamma correction with γ = 0.5.

```python
import cv2 as cv
import numpy as np

gamma = 0.5
f = cv.imread('media\\runway.png', cv.IMREAD_GRAYSCALE)/255.0

cv.namedWindow('image', cv.WINDOW_AUTOSIZE)
cv.imshow('image', f)
cv.waitKey(0)

g = f**gamma

cv.imshow('image', g)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import cv2 as cv
import numpy as np

gamma = 0.5
f = cv.imread('media\\runway.png', cv.IMREAD_GRAYSCALE)

cv.namedWindow('image', cv.WINDOW_AUTOSIZE)
cv.imshow('image', f)
cv.waitKey(0)

t = np.array([(i/255.0)**gamma*255 for i in range(256)], dtype='uint8')
g = cv.LUT(f, t) # f : image, t : transform

cv.imshow('image', g)
cv.waitKey(0)
cv.destroyAllWindows()
```

(c) Contrast Stretching (linear piecewise transformation)

```python
import cv2 as cv
import numpy as np

img = cv.imread('media\\runway.png', cv.IMREAD_GRAYSCALE)
r = img/255.0 # normalize to [0,1]

r1 = 0.2
r2 = 0.8

s = np.zeros_like(r) # creating an empty output image

mask1 = r < r1
s[mask1] = 0

mask2 = (r >= r1) & (r <= r2)
s[mask2] = (r[mask2] - r1) / (r2 - r1)

mask3 = r > r2
s[mask3] = 1

s = (s * 255).astype('uint8') # scale back to [0,255]
cv.imshow('Original Image', img)
cv.imshow('Contrast Stretched Image', s)
cv.waitKey(0)
cv.destroyAllWindows()
```
