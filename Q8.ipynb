```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# 1. Load the Noisy Image
# Based on your file list, the image is likely named 'emma_salt_pepper.jpg'
img = cv.imread('media/emma_salt_pepper.jpg', cv.IMREAD_GRAYSCALE)

if img is None:
    print("Error: Could not load image.")
else:
    gaussian_result = cv.GaussianBlur(img, (5, 5), 2)
    median_result = cv.medianBlur(img, 5)

    plt.figure(figsize=(15, 6))

    # Original
    plt.subplot(1, 3, 1)
    plt.imshow(img, cmap='gray')
    plt.title("Original (Salt & Pepper Noise)")
    plt.axis('off')

    # Gaussian Result
    plt.subplot(1, 3, 2)
    plt.imshow(gaussian_result, cmap='gray')
    plt.title("(a) Gaussian Smoothing")
    plt.axis('off')

    # Median Result
    plt.subplot(1, 3, 3)
    plt.imshow(median_result, cmap='gray')
    plt.title("(b) Median Filtering")
    plt.axis('off')

    plt.show()
```
