```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img_bgr = cv.imread('media/runway.png')

if img_bgr is None:
    print("Error: Image not found.")
else:
    img_rgb = cv.cvtColor(img_bgr, cv.COLOR_BGR2RGB)

    kernel_sharpening = np.array([[ 0, -1,  0],
                                  [-1,  5, -1],
                                  [ 0, -1,  0]])


    img_sharpened = cv.filter2D(img_rgb, -1, kernel_sharpening)

    plt.figure(figsize=(15, 8))

    plt.subplot(1, 2, 1)
    plt.imshow(img_rgb)
    plt.title("Original Image")
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(img_sharpened)
    plt.title("Sharpened Image")
    plt.axis('off')

    plt.show()
```
